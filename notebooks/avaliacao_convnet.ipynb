```python
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df_y = pd.read_csv("../execucao/convnet_test_preds.csv")
df_loss = pd.read_csv("../execucao/convnet_loss.csv")
```

```python
y = df_y.y_true.values
y_pred = df_y.y_pred.values
```

```python
confusion_matrix(y, y_pred)
```

```python
print(classification_report(y, y_pred))
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
df_loss[["train_loss", "val_loss"]].rolling(5).mean().plot(ax=ax, color=['red', 'blue'])
df_loss[["train_loss", "val_loss"]].plot(ax=ax, alpha=.2, color=['red', 'blue'])
plt.show()
```
